```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('10Year Swap Rate.csv')
data
```

```python
title = data.columns[1:].values
```

```python
data
data = np.array(data)
data
```

```python
currency = 'USD'
index = 'US swap'


dat = []
for i in range(len(data)):
    dat_day = np.zeros(shape = (len(data[0])-1+4,2)).astype('O')
    dat_day[:4,0] = ['CCY','INDEX','DATE','ID']
    dat_day[:4,1] = [currency, index, data[:,0][i], 'TEAM3']
    dat_day[4:,0] = title
    dat_day[4:,1] = data[i][1:]
    dat.append(dat_day)

    
   
        
        
    
    
    
```

```python
dat
```

```python
import codecs



f = codecs.open("test.txt",'w','utf-8')


#f.write(str(list))
for i in dat:
    f.write(str(i)+'\r\n')  #\r\n为换行符

f.close()
```
